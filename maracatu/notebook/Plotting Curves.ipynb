```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-1, 1)
y = np.sin(x * 2 * np.pi ) 
df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

## 1. Import maracatu

```python
import maracatu as m
```

## 2. Create a dictionary with parameters

```python
par_base = {
    'title': 'Plotting curves with Maracatu',
    'xlabel': 'X label',
    'ylabel': 'Y label',
    'alpha': 0.8,
    'grid': True, 
    'fig.tight_layout': True,
}
```

## 3. Plot 

```python
plt_curves = m.Curves()
fig, ax, par_plot = plt_curves.plot(df, par_base)
```
