```python
import numpy as np
import pandas as pd
```

```python
x = np.random.multivariate_normal([0,0], [[1,.6],[.6,1]], 1000)
df = pd.DataFrame({'X': x[:,0], 'Y': x[:,1]})
df.head()
```

## 1. Import maracatu

```python
import maracatu as m
```

## 2. Create a dictionary with parameters

```python
par_base = {
    'title': 'Plotting Scatter with Maracatu',
    'xlabel': 'X label',
    'ylabel': 'Y label',
    'alpha': 0.4,
    'grid': True, 
    'fig.tight_layout': True,
}
```

## 3. Plot 

```python
plt_scatter = m.Scatter()
fig, ax, par_plot = plt_scatter.plot(df, par_base)
```
